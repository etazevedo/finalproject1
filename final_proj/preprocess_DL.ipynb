```python
from PIL import Image
from docx2pdf import convert
import cv2
import pytesseract as tess
import numpy as np
import docx
import matplotlib.pyplot as plt
import requests
```

```python
image_url = 'http://casacomum.org/aebdocs/05/05755/05755.019.04954/d2/05755.019.04954_p0002_id000617765_D2.png'

img_data = requests.get(image_url).content
with open('image_name.jpg', 'wb') as handler:
    handler.write(img_data)

 
im_file = 'image_name.jpg'
im = cv2.imread(im_file)
im_original = im.copy()
#im.show()
display(im)
```

```python
def noise_removal(image):

    kernel = np.ones((1, 1), np.uint8)
    image = cv2.dilate(image, kernel, iterations=1)
    kernel = np.ones((1, 1), np.uint8)
    image = cv2.erode(image, kernel, iterations=1)
    image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    image = cv2.medianBlur(image, 3)
    return (image)
```

```python
noise_removal(im)
```

```python

gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

kernal = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 13))
cv2.imwrite("temp/image_name.jpg", kernal)

dilate = cv2.dilate(thresh, kernal, iterations=1)
cv2.imwrite("temp/image_dilate.png", dilate)


ROI_number = 0
cnts = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]
cnts = sorted(cnts, key=lambda x: cv2.boundingRect(x)[0])
for c in cnts:
    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(im, (x, y), (x + w, y + h), (0,0,255), 2)
    
    

#cv2.imshow('image', im)
#cv2.imshow('Thresh',thresh)
#cv2.imshow('Kernal',kernal)
#cv2.imshow('Dilate',dilate)
#cv2.waitKey()
```
